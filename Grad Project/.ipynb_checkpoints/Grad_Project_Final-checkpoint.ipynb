## 1.1 kepler_exoplanet_search.csv

The first dataset (the Kepler dataset) in this study is collected by NASA from the Kepler Space Observatory. In 2009, NASA launched the Kepler Mission, an effort to discover exoplanets (planets outside the Solar System), with the goal of finding potential habitable planets for human beings. The mission lasted for over nine years with remarkable legacies—a total number of 9,564 potential exoplanets are contained in the Kepler dataset, each associated with features that indicate the characteristics of the exoplanet. Among these features, there is a categorical target variable, koi_disposition, with three possible values, “CONFIRMED”, “CANDIDATE”, and “FALSE POSITIVE”. If an exoplanet is “CONFIRMED”, its existence is confirmed, and it is associated with a name recorded by kepler_name; if an exoplanet is “CANDIDATE”, its existence has not been proven yet; if an exoplanet is “FALSE POSITIVE”, it is proved to be non-existent. There are totally 2,358 confirmed exoplanets, 2,366 candidate exoplanets, and the rest 4,840 exoplanets are false positive. This dataset is used for supervised classification.

# Part 2. Data Cleaning

The first step of data cleaning is to calculate the proportion of empty entries in each column. Columns with a large proportion of empty data are excluded. Then columns with a small fraction of empty data are manually selected and the empty entries in the preserved columns are filled with proper values. Finally, some data are dropped if they contain empty values after data cleaning. 

## 2.1 Calculate the proportion of empty entries in each column

```python
kepler_exoplanet_search_empty_entries_proportion = defaultdict(list)
kepler_planetary_system_composite_empty_entries_proportion = defaultdict(list)
```

```python
def empty_proportion(column_name, dataframe):
    if dataframe == 'kepler_planetary_system_composite':
        num_empty_entries = kepler_planetary_system_composite[column_name].isna().sum(axis=0)
        fraction = num_empty_entries/4375
        kepler_planetary_system_composite_empty_entries_proportion[fraction].append(column_name)
    else:
        num_empty_entries = kepler_exoplanet_search[column_name].isna().sum(axis=0)
        fraction = num_empty_entries/9564
        kepler_exoplanet_search_empty_entries_proportion[fraction].append(column_name)
```

```python
for column in kepler_exoplanet_search.columns:
    empty_proportion(column, 'kepler_exoplanet_search')
for column in kepler_planetary_system_composite.columns:
    empty_proportion(column, 'kepler_planetary_system_composite')
```

## 2.2 Remove columns with a large proportion of empty entries

```python
# For the first dataset, just drop columns with all empty data
kepler_exoplanet_search_empty_columns = kepler_exoplanet_search_empty_entries_proportion[1]
kepler_exoplanet_search.drop(columns=kepler_exoplanet_search_empty_columns, inplace=True)
```

```python
# For the second dataset, set a threshold
threshold = 0.01
kepler_planetary_system_composite_empty_columns = []
for key in sorted(kepler_planetary_system_composite_empty_entries_proportion.keys(), reverse=True):
    if key < 0.01: break
    else: kepler_planetary_system_composite_empty_columns.extend(kepler_planetary_system_composite_empty_entries_proportion[key])
kepler_planetary_system_composite.drop(columns=kepler_planetary_system_composite_empty_columns, inplace=True)
```

## 2.3 Fill with proper values for columns

```python
# We manually select some columns to drop and fill other empty entries with 0s
columns_to_drop = ['koi_datalink_dvr', 'koi_gmag', 'koi_rmag', 'koi_imag', 'koi_zmag', 'koi_jmag', 'koi_hmag', 'koi_kmag']
kepler_exoplanet_search.drop(columns=columns_to_drop, inplace=True)
kepler_exoplanet_search.dropna(axis=0, subset=['koi_kepmag'], inplace=True)
kepler_exoplanet_search.fillna(0, inplace=True)
columns_to_drop = ['sy_mnum', 'st_masslim']
kepler_planetary_system_composite.drop(columns=columns_to_drop, inplace=True)
kepler_planetary_system_composite.dropna(axis=0, subset=list(kepler_planetary_system_composite.columns), inplace=True)
kepler_planetary_system_composite.fillna(0, inplace=True)
```

```python
# Take log scale period
kepler_exoplanet_search['koi_period'] = np.log(kepler_exoplanet_search['koi_period'])
kepler_exoplanet_search['koi_time0'] = np.log(kepler_exoplanet_search['koi_time0'])
# Change values of the target variable
kepler_exoplanet_search.replace(to_replace={'CONFIRMED':1, 'CANDIDATE':0, 'FALSE POSITIVE':-1}, inplace=True)
```

# Part 4. Feature Selection

Based on the correlation heatmaps, we excluded variables that have high correlation coefficients with one or more other variables.

## 4.1 Feature selection of the Kepler dataset

## 4.2 Feature selection of the confirmed exoplanets dataset

```python
Exoplanets_dataset_classification_columns = ['sy_pnum', 'cb_flag', 'pul_flag', 'ptv_flag', 'ast_flag', 'dkin_flag',
                                             'pl_radj', 'pl_bmassj', 'ttv_flag', 'st_mass', 'ra', 'dec', 'sy_dist',
                                             'st_nphot', 'st_nrvc', 'pl_nespec']
Exoplanets_dataset_identification_columns = ['pl_name']
Exoplanets_all_columns = Exoplanets_dataset_identification_columns+Exoplanets_dataset_classification_columns
Exoplanets_features = kepler_planetary_system_composite[Exoplanets_all_columns]
Exoplanets_features.head()
```

# Part 5. Model Training, Test, and Evaluation

```python
model_data = Kepler_features.copy()
features = model_data[Kepler_dataset_classification_columns]
targets = model_data[Kepler_dataset_target]
train_ratio = 0.8
X_train, X_test, y_train, y_test = features[0:int(len(features)*train_ratio)], features[int(len(features)*train_ratio):], targets[0:int(len(targets)*train_ratio)], targets[int(len(targets)*train_ratio):] 
```

## 5.1 Supervised learning of the Kepler dataset

### 5.1.1 Classification tree

```python
test_err_tree = dict()
for i in range(2, 100, 1): 
    for j in range(1, 20):
        tree_model = tree.DecisionTreeClassifier(min_samples_split=i, max_depth=j, random_state=42) 
        tree_model.fit(X_train, y_train)
        test_err_tree[(i,j)] = accuracy_score(tree_model.predict(X_test), y_test)
```

```python
print('Highest accuracy:', max(test_err_tree.values()))
optimal_i, optimal_j = max(test_err_tree, key=test_err_tree.get)
print('The minimum number of samples required to split an internal node:', optimal_i)
print('The maximum depth of the tree:', optimal_j)
```

```python
# Get the optimal tree model and visualize it
optimal_tree_model = tree.DecisionTreeClassifier(min_samples_split=optimal_i, max_depth=optimal_j, random_state=42); 
optimal_tree_model.fit(X_train, y_train);
tree.plot_tree(optimal_tree_model);
```

```python
# Export the decision tree
tree.export_graphviz(optimal_tree_model,
                     out_file="tree.dot",
                     feature_names = Kepler_dataset_classification_columns,
                     filled = True,
                     rounded=True,  
                     special_characters=True)
```

```python
# Visualize feature importances
tree_importances = optimal_tree_model.feature_importances_
tree_importances = pd.Series(tree_importances, index=Kepler_dataset_classification_columns).sort_values(ascending=False)
fig, ax = plt.subplots();
tree_importances.plot.bar(ax=ax);
ax.set_title("Feature Importances of the Kepler Dataset", fontsize=32);
plt.xlabel(xlabel="Features", fontsize=24);
plt.ylabel(ylabel="Gini Importance", fontsize=24);
plt.xticks(size=18);
plt.yticks(size=18);
for p in ax.patches:
    ax.annotate(format(p.get_height(), ".5f"), (p.get_x()+0.25, p.get_height()-0.01), ha='center', va='top', color='white', size=18);
fig.tight_layout();
```

### 5.1.2 Random forest

```python
test_err_rf = dict()
oob_err_rf = []

y_train_rf = pd.DataFrame()
y_train_rf['FALSE POSITIVE'] = y_train.replace({-1:1, 1:0, 0:0})['koi_disposition']
y_train_rf['CANDIDATE'] = y_train.replace({-1:0, 1:0, 0:1})['koi_disposition']
y_train_rf['CONFIRMED'] = y_train.replace({-1:0, 0:0})['koi_disposition']

for i in range(20, 1000, 20):
    f_model = rf(n_estimators=i, oob_score=True, random_state=42)
    f_model.fit(X_train, y_train_rf)
    one_hot = f_model.predict(X_test)
    predicted_labels = [np.where(label == np.max(label))[0][0]-1 for label in one_hot]
    test_err_rf[i] = accuracy_score(predicted_labels, y_test)
    oob_err_rf.append(f_model.oob_score_)
```

```python
print('Highest accuracy:', max(test_err_rf.values()))
i = max(test_err_rf, key=test_err_rf.get)
print('The number of trees in the forest:', i)
optimal_rf_model = rf(n_estimators=i, oob_score=True, random_state=42);
optimal_rf_model.fit(X_train, y_train_rf);
```

```python
# Visualize OOB Score
plt.plot(range(20,1000,20), np.array(oob_err_rf));
plt.title("OOB Score of the Random Forest Model", fontsize=32);
plt.xlabel("Number of Trees", size=18);
plt.ylabel("OOB Score", size=18);
plt.show();
```

### 5.1.3 Naive Bayes

```python
scaler = MinMaxScaler()
X_train_nb = scaler.fit_transform(X_train)
X_test_nb = scaler.fit_transform(X_test)
nb_model = nb(fit_prior=True, class_prior=[4840/9564, 2366/9564, 2358/9564])
nb_model.fit(X_train_nb, y_train)
print('Accuracy:', accuracy_score(nb_model.predict(X_test_nb), y_test))
```

### 5.1.4 Neural network

```python
test_err_mlp = dict()

activations = ['logistic', 'tanh', 'relu']
solvers = ['lbfgs', 'sgd', 'adam']
for f in activations:
    for s in solvers:
        for layer_size in range(10, 100, 5):
            for lr in np.arange(0.001, 0.01, 0.001):
                nn_model = mlp(activation=f, solver=s, learning_rate_init=lr, random_state=42, hidden_layer_sizes=layer_size)
                nn_model.fit(X_train, y_train_rf)
                one_hot = nn_model.predict(X_test)
                predicted_labels = [np.where(label == np.max(label))[0][0]-1 for label in one_hot]
                test_err_mlp[(f,s,layer_size,lr)] = accuracy_score(predicted_labels, y_test)
```

```python
print('Highest accuracy:', max(test_err_mlp.values()))
optimal_f, optimal_s, optimal_layer_size, optimal_lr = max(test_err_mlp, key=test_err_mlp.get)
print('Activation function:', optimal_f)
print('Solver for weight optimization:', optimal_s)
print('Layer size:', optimal_layer_size)
print('Learning rate:', optimal_lr)
optimal_nn_model = mlp(activation=optimal_f, solver=optimal_s, learning_rate_init=optimal_lr, random_state=42, 
                       hidden_layer_sizes=optimal_layer_size);
optimal_nn_model.fit(X_train, y_train_rf);
```

### 5.1.5 Cross validation

```python
k = 10
kf = KFold(n_splits=k, shuffle=True, random_state=42)

best_tree = tree.DecisionTreeClassifier(min_samples_split=53, max_depth=7, random_state=42);
best_random_forest = rf(n_estimators=40, oob_score=True, random_state=42);
best_naive_bayes = nb(fit_prior=True, class_prior=[4840/9564, 2366/9564, 2358/9564]);
best_neural_network = mlp(activation='tanh', solver='adam', learning_rate_init=0.003, random_state=42, hidden_layer_sizes=25);

best_models = [(best_tree, 0, 0, 'Tree'), (best_rf, 1, 0, 'Random Forest'), (best_nb, 0, 1, 'Naive Bayes'), (best_nn, 1, 0, 'Neural Network')]
acc_score = defaultdict(list)

for best_model, one_hot, is_nb, name in best_models:
    for train_index , test_index in kf.split(features):
        X_train, X_test = features.iloc[train_index,:], features.iloc[test_index,:]
        y_train, y_test = targets.iloc[train_index], targets.iloc[test_index]
        
        if is_nb: 
            scaler = MinMaxScaler()
            X_train_nb = scaler.fit_transform(X_train)
            X_test_nb = scaler.fit_transform(X_test)
            best_model.fit(X_train_nb, y_train)
            pred_values = best_model.predict(X_test_nb)
        
        else:
            best_model.fit(X_train, y_train)
            pred_values = best_model.predict(X_test)
     
        if one_hot:
            predicted_labels = [np.where(label == np.max(label))[0][0]-1 for label in pred_values]
            acc = accuracy_score(predicted_labels, y_test)
            acc_score[name].append(acc)
        
        else:
            acc = accuracy_score(pred_values, y_test)
            acc_score[name].append(acc)
```

```python
for key in acc_score:
    print('Avg accuracy of', key, ':', np.mean(acc_score[key]))
```
