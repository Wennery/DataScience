### Exploratory Data Analysis - kepler_exoplanet_search

Exoplanet Archive Information is the master index of which exoplanets have been "found", either as CONFIRMED, CANDIDATE, FALSE POSITIVE.

When an exoplanet gets "found" it is also given a human-friendly name.

#### KOI Dispositions - CANDIDATE / CONFIRMED / FALSE POSITIVE

```python
koi_disposition_counts = (
    pd.DataFrame(
        exoplanet_filtered
            .groupby('koi_disposition')
            .size()
            .map(lambda count: {
                "count":      count, 
                "percentage": round(count/exoplanet_filtered.shape[0] * 100, 1)
            })
            .to_dict()
    ).transpose()
)
koi_disposition_counts
```

```python
def pie_label(pct, allvals):
    absolute = int(pct/100.*np.sum(allvals))
    return "{:d}\n({:.1f}%)".format(absolute, pct)    

plt.pie(
    koi_disposition_counts['count'], 
    labels=koi_disposition_counts.index,     
    explode=(0,0.1,0,), 
    #     autopct='%1.1f%%',
    autopct=lambda pct: pie_label(pct, koi_disposition_counts['count']),    
    shadow=True, 
)
plt.title("KOI Dispositions")
plt.show()

koi_disposition_counts['count']
```

### Exploring kepler_planetary_system_composite

```python
kepler_planetary_system_composite['goldilocks_temp'] = ((273.2 <= kepler_planetary_system_composite['pl_eqt']) & (kepler_planetary_system_composite['pl_eqt'] <= 373.2))
goldilocks_counts = {}
goldilocks_counts['temp'] = {
    "too cold":   kepler_planetary_system_composite.query('pl_eqt         <= 273.2').shape[0],
    "just right": kepler_planetary_system_composite.query('goldilocks_temp == True' ).shape[0],
    "too hot":    kepler_planetary_system_composite.query('pl_eqt         >= 373.2').shape[0],
}
for key, value in goldilocks_counts['temp'].items():
    print( 'Exoplanets that are %-10s: %4d (%5.2f%%)' % ( key, value, 100*value/kepler_planetary_system_composite.shape[0] ) )    
```

```python
kepler_planetary_system_composite['goldilocks_size'] = ((0.8 <= kepler_planetary_system_composite['pl_rade']) & (kepler_planetary_system_composite['pl_rade'] <= 1.7))
goldilocks_counts['size'] = {
    "too small":  kepler_planetary_system_composite.query('pl_rade        <= 0.8' ).shape[0],
    "just right": kepler_planetary_system_composite.query('goldilocks_size == True').shape[0],
    "too big":    kepler_planetary_system_composite.query('pl_rade        >= 1.7' ).shape[0],
}
for key, value in goldilocks_counts['size'].items():
    print( 'Exoplanets that are %-10s: %4d (%5.2f%%)' % ( key, value, 100*value/kepler_planetary_system_composite.shape[0] ) )    
```

```python
kepler_planetary_system_composite['goldilocks'] = ((kepler_planetary_system_composite['goldilocks_temp'] == True) & (kepler_planetary_system_composite['goldilocks_size'] == True))
goldilocks_counts['combined'] = {
    "just right": kepler_planetary_system_composite.query('goldilocks==True').shape[0]
}

for key in goldilocks_counts.keys():
    value = goldilocks_counts[key]['just right']
    print( 'Exoplanets that are "just right" %-10s: %4d (%5.2f%%)' % ( key, value, 100*value/kepler_planetary_system_composite.shape[0] ) ) 
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(   "Number of potentually habitable exoplanets: " + str(kepler_planetary_system_composite.query('goldilocks==True').shape[0]) )
    print(   "Names of potentually habitable exoplanets: " + ", ".join( kepler_planetary_system_composite.query('goldilocks==True')['pl_name'].tolist()) )    
    display( kepler_planetary_system_composite.query('goldilocks==True') )
```

```python
df = kepler_planetary_system_composite.query('goldilocks==True')
df = df.drop(1078) # index 1078 has a NaN pl_orbsmax value
df['KMeans_StarType'] = sklearn.cluster.KMeans(n_clusters=4).fit_predict(df[['pl_orbsmax','st_mass']])
df = df.reset_index()
```

```python
plt.figure(figsize=(12,8))
plot = sns.scatterplot(
    data=df,        
    x="pl_orbsmax",
    y="st_mass",

    size="pl_rade", sizes=(20,400),    
    hue="KMeans_StarType", palette="Blues",
#     hue="koi_teq", palette="RdBu_r",
)
for line in range(0, df.shape[0]):
     plot.text(
         df['pl_orbsmax'][line]+0.005, 
         df['st_mass'][line], 
         df['pl_name'][line], 
         horizontalalignment='left', 
         size='medium', 
         color='black', 
         # weight='semibold'
     )

plt.title('Confirmed Goldilocks Exoplanets by StarType')
plt.xlabel('Semi Major Axis / Orbital Distance (AU)')
plt.ylabel('Stellar Mass (solar mass)')
display()
```

```python
df['KMeans_PlanetType'] = sklearn.cluster.KMeans(n_clusters=6).fit_predict(df[['st_met','pl_rade']])
```

```python
plt.figure(figsize=(12,8))
plot = sns.scatterplot(
    data=df,        
    x="st_met",
    y="pl_rade",

    size="pl_rade", sizes=(20,400),    
    hue="KMeans_PlanetType", palette="Accent",
)
for line in range(0,df.shape[0]):
     plot.text(
         df['st_met'][line]+0.005, 
         df['pl_rade'][line], 
         df['pl_name'][line], 
         horizontalalignment='left', 
         size='medium', 
         color='black', 
         # weight='semibold'
     )

plt.title('Confirmed Goldilocks Exoplanets by PlanetType')
plt.xlabel('Stellar Metallicity')
plt.ylabel('Planetary Radius (Earth radii)')
display()
```
