### Biuld Machine Learning Model to Predict "Very Likely to Be Confirmed" Exoplanets from "CANDIDATE"

```python
model_data = exoplanet_filtered_finished.copy()
model_data.head()
```

```python
model_data.info()
```

#### Removing Name-related and Non-value Variables

```python
del model_data['kepoi_name']
del model_data['kepler_name']
del model_data['koi_datalink_dvr']
```

```python
model_data.info()
```

#### Missing Value Imputation
From the above output, we find that the following variables need missing value imputation:

```python
missing_col = ('koi_period_err1', 'koi_period_err2', 'koi_time0bk_err1', 'koi_time0bk_err2', 'koi_time0_err1', 'koi_time0_err2', 
               'koi_kepmag', 'koi_gmag', 'koi_rmag', 'koi_imag', 'koi_zmag', 'koi_jmag', 'koi_hmag', 'koi_kmag')
```

```python
for col in missing_col:
    model_data[col] = model_data[col].fillna(model_data[col].mean())
```

```python
model_data.info()
```

```python
model_data = pd.get_dummies(model_data)
model_data
```

Deleting some dummy variables created to avoid multi-collinearity.

```python
del model_data['koi_disposition_FALSE POSITIVE']
del model_data['koi_pdisposition_CANDIDATE']
del model_data['koi_pdisposition_FALSE POSITIVE']
```

```python
corr = model_data.corr()
corr
```

```python
del model_data['kepid']
```

```python
model_data.shape
```

Our target variable has 3 categories, therefore we need to make a variable which takes values 0, 1, and 2 to represent FALSE POSITIVE, CANDIDATE, and CONFIRMED respectively.

```python
model_data['koi_disposition'] = [0]*9564
```

```python
for i in range(0, 9564):
    if(model_data.at[i, 'koi_disposition_CANDIDATE'] == 1):
        model_data.at[i, 'koi_disposition'] = 1
    elif(model_data.at[i, 'koi_disposition_CONFIRMED'] == 1):
        model_data.at[i, 'koi_disposition'] = 2
    elif(model_data.at[i, 'koi_disposition_CANDIDATE'] == 0 and model_data.loc[i, 'koi_disposition_CONFIRMED'] == 0):
        model_data.loc[i, 'koi_disposition'] = 0
```

```python
model_data
```

```python
del model_data['koi_disposition_CANDIDATE']
del model_data['koi_disposition_CONFIRMED']
```

#### Split Train and Test Sets

```python
split_ratio = 0.8
model_data.shape[0]*split_ratio
```

```python
train_data = model_data.loc[:7651, :]
test_data = model_data.loc[7652:, :]
```

```python
x_train = train_data.loc[:, train_data.columns != 'koi_disposition']
y_train = train_data['koi_disposition']
x_test = test_data.loc[:, test_data.columns != 'koi_disposition']
y_test = test_data['koi_disposition']
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape, model_data.shape
```

#### Model Biulding - Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
ran_for = RandomForestClassifier(n_estimators = 10000, n_jobs = -1, random_state = 0)
```

```python
ran_for.fit(x_train, y_train)
```

```python
ran_train_predict = ran_for.predict(x_train)
ran_test_predict = ran_for.predict(x_test)
```

```python
ran_train_acc = accuracy_score(y_train, ran_train_predict)
ran_test_acc = accuracy_score(y_test, ran_test_predict)
ran_train_acc, ran_test_acc
```

## Our model gives a good accuracy (96.5%)!

#### Which "CANDIDATES" in the Test Set is Now Classified As "Confirmed" by Our Model?

```python
ran_test_predict_list = ran_test_predict.tolist()
```

```python
new_confirmed_index = {}
count = 0
for i in range(9564-ran_test_predict.shape[0], 9564):
    if ran_test_predict_list[9564-1-i] == 2:
        new_confirmed_index[count] = i
        count = count + 1
    
```

```python
new_confirmed_index = list(new_confirmed_index.values())
```

```python
new_confirmed_index_array = np.asarray(new_confirmed_index)
new_confirmed_index_array
```

```python
ex_candidate_planet = exoplanet_filtered_finished[exoplanet_filtered_finished['koi_disposition']=='CANDIDATE']
```

```python
boolean_series = ex_candidate_planet.index.isin(new_confirmed_index_array)
```

```python
ex_candidate_planet_now_confirmed = ex_confirmed_planet[boolean_series]
ex_candidate_planet_now_confirmed
```

## Conclusion: 
### 1. Only 9 "CANDIDATES" in the test set is now classified as "CONFIRMED" by our model. These 9 exoplanets are "very likely to be confirmed" "CANDIDATES". 
### 2. We note that none of these 9 planets have any false positive flags (koi_fpflag_xx). This makes sense.
### 3. Since there are only 17 goldilocks exoplanets among 2358 "Confirmed" exoplanets, it is very unlikely to discover a new goldilocks exoplanet from only 9 "very likely to be confirmed" "CANDIDATES".
